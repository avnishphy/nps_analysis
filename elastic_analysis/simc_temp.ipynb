```python
import os
import re
import glob

# Directory containing the report files
report_dir = "/lustre24/expphy/volatile/hallc/nps/singhav/REPORT_OUTPUT/COIN/"

# Function to parse a single report file and extract desired parameters
def parse_report(report_file):
    params = {
        'BCM2 Current': None,
        'BCM4A Current': None,
        'BCM2 Charge': None,
        'BCM4A Charge': None,
        'Total Live Time (EDTM)': None,
        'E SING FID TRACK EFFIC': None,
        '3_of_4 EFF': None,
        'Pre-Scaled Ps1 HMS Computer Live Time': [],
        'Pre-Scaled Ps2 HMS Computer Live Time': [],
        'Pre-Scaled Ps3 HMS Computer Live Time': [],
        'Pre-Scaled Ps4 HMS Computer Live Time': [],
        'Pre-Scaled Ps5 HMS Computer Live Time': [],
        'Pre-Scaled Ps6 HMS Computer Live Time': []
    }

    with open(report_file, 'r') as f:
        for line in f:
            if 'BCM2 Current' in line:
                # params['BCM2 Current'] = float(re.findall(r"[-+]?\d*\.\d+|\d+", line)[0])
                params['BCM2 Charge'] = re.findall(r"[-+]?\d*\.\d+\s*[a-zA-Z]+", line)[0]
            elif 'BCM4A Current' in line:
                params['BCM4A Current'] = float(re.findall(r"[-+]?\d*\.\d+|\d+", line)[0])
            elif 'BCM2 Charge' in line:
                params['BCM2 Charge'] = float(re.findall(r"[-+]?\d*\.\d+|\d+", line)[0])
            elif 'BCM4A Charge' in line:
                params['BCM4A Charge'] = float(re.findall(r"[-+]?\d*\.\d+|\d+", line)[0])
            elif 'Total Live Time (EDTM)' in line:
                params['Total Live Time (EDTM)'] = float(re.findall(r"[-+]?\d*\.\d+|\d+", line)[0])
            elif 'E SING FID TRACK EFFIC' in line:
                params['E SING FID TRACK EFFIC'] = float(re.findall(r"[-+]?\d*\.\d+|\d+", line)[0])
            elif '3_of_4 EFF' in line:
                params['3_of_4 EFF'] = float(re.findall(r"[-+]?\d*\.\d+|\d+", line)[0])
            elif 'Pre-Scaled Ps1 HMS Computer Live Time' in line:
                params['Pre-Scaled Ps1 HMS Computer Live Time'].append(float(re.findall(r"[-+]?\d*\.\d+|\d+", line)[0]))
            elif 'Pre-Scaled Ps2 HMS Computer Live Time' in line:
                params['Pre-Scaled Ps2 HMS Computer Live Time'].append(float(re.findall(r"[-+]?\d*\.\d+|\d+", line)[0]))
            elif 'Pre-Scaled Ps3 HMS Computer Live Time' in line:
                params['Pre-Scaled Ps3 HMS Computer Live Time'].append(float(re.findall(r"[-+]?\d*\.\d+|\d+", line)[0]))
            elif 'Pre-Scaled Ps4 HMS Computer Live Time' in line:
                params['Pre-Scaled Ps4 HMS Computer Live Time'].append(float(re.findall(r"[-+]?\d*\.\d+|\d+", line)[0]))
            elif 'Pre-Scaled Ps5 HMS Computer Live Time' in line:
                params['Pre-Scaled Ps5 HMS Computer Live Time'].append(float(re.findall(r"[-+]?\d*\.\d+|\d+", line)[0]))
            elif 'Pre-Scaled Ps6 HMS Computer Live Time' in line:
                params['Pre-Scaled Ps6 HMS Computer Live Time'].append(float(re.findall(r"[-+]?\d*\.\d+|\d+", line)[0]))
    
    return params

# Function to extract run number from a filename
def extract_run_number(file):
    match = re.search(r'coin_NPS_HMS_report_(\d+)\.report', file)
    return int(match.group(1)) if match else None

# Main function to process runs and handle ranges
def process_runs(run_numbers):
    for run_number in run_numbers:
        # Find the report file for the run number
        report_files = glob.glob(os.path.join(report_dir, f"coin_NPS_HMS_report_{run_number}*.report"))
        if report_files:
            report_file = report_files[0]
            print(f"\nProcessing file: {os.path.basename(report_file)}")

            # Extract parameters
            params = parse_report(report_file)

            # Print results
            for param, value in params.items():
                if value is not None:
                    print(f"{param}: {value}")
        else:
            print(f"No report file found for run number {run_number}")

# Function to parse input for run numbers, including ranges
def parse_run_numbers(run_input):
    run_numbers = set()
    for part in run_input.split(','):
        if '-' in part:
            start, end = map(int, part.split('-'))
            run_numbers.update(range(start, end + 1))
        else:
            run_numbers.add(int(part))
    return sorted(run_numbers)

# Example usage
if __name__ == '__main__':
    # Input: run numbers (can include ranges like '100-105,110')
    run_input = input("Enter run numbers (e.g., '100-105,110'): ")
    run_numbers = parse_run_numbers(run_input)
    process_runs(run_numbers)
```
